```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = {
    'Bread': [1,1,1,1,0],
    'Butter': [1, 1, 0, 0, 1],
    'Jam': [1, 0,0,0,0],
    'Milk': [1,1,0,1,1],
    'Juice': [0,0, 1,1,1],
    'Curd': [0,0,1,0,0]
}
df = pd.DataFrame(data)
print(f"The dataset is: \n{df}")
```

```python
freqItems = apriori(
    df, min_support=0.5,
    use_colnames=True,
    max_len=2
)

print(f"The Frequent Items are: \n{freqItems}")
```

```python
rules = association_rules(
    freqItems,
    metric='confidence',
    min_threshold=0.75
)
print(rules)
```
